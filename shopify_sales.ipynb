```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
december23sales = pd.read_csv('/content/sales_2023-12-01_2023-12-31 (1).csv')

# Convert 'order_date' to datetime format
december23sales['day'] = pd.to_datetime(december23sales['day'])

# Filter data for specific products (replace with actual product names)
specific_products = ["REPLICA JERSEY ALT FAUXBACK PP22", "TEE ROSCOE ABCD ", "REPLICA JERSEY WHITE PP22",

]
filtered_data = december23sales[december23sales['product_title'].isin(specific_products)]

# Aggregate sales by date for each product
time_series_data = (
    filtered_data.groupby(['day', 'product_title'])
    .agg({'net_sales': 'sum'})
    .reset_index()
)

# Pivot data to create columns for each product
time_series_pivot = time_series_data.pivot(index='day', columns='product_title', values='net_sales').fillna(0)

# Plot the time series
plt.figure(figsize=(12, 6))
for product in specific_products:
    if product in time_series_pivot.columns:
        plt.plot(time_series_pivot.index, time_series_pivot[product], label=product)

# Customize the plot
plt.title('Time Series Analysis of Sales for Specific Products')
plt.xlabel('Date')
plt.ylabel('net_sales')
plt.legend(title="Products")
plt.grid()
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# @title day vs product_id

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['day']
  ys = series['product_id']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = december23sales.sort_values('day', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('product_type')):
  _plot_series(series, series_name, i)
  fig.legend(title='product_type', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('day')
_ = plt.ylabel('product_id')
```

```python

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['day']
  ys = series['product_id']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = december22sales.sort_values('day', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('product_type')):
  _plot_series(series, series_name, i)
  fig.legend(title='product_type', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('day')
_ = plt.ylabel('product_id')
```
